```python
import os
import requests
import logging
from dotenv import load_dotenv
from concurrent.futures import ThreadPoolExecutor

# Load environment variables from .env file
load_dotenv()

# Set up logging
logging.basicConfig(level=logging.INFO)

# CoinMarketCap API URL
coinmarketcap_api_url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'

# Replace 'YOUR_COINMARKETCAP_API_KEY' with your actual CoinMarketCap API key
coinmarketcap_api_key = os.getenv('COINMARKETCAP_API_KEY')

# Binance Testnet API URL
binance_api_url = 'https://api.binance.com/api/v3/ticker/price'

# Replace 'YOUR_BINANCE_API_KEY' with your actual Binance Testnet API key
binance_api_key = os.getenv('BINANCE_API_KEY')

# CoinGecko API URL
coingecko_symbols_url = 'https://api.coingecko.com/api/v3/coins/list'
coingecko_api_url = 'https://api.coingecko.com/api/v3/simple/price'

# No API key is required for CoinGecko

# Fetch supported symbols directly from CoinGecko
try:
    response = requests.get(coingecko_symbols_url)
    response.raise_for_status()
    supported_symbols = [symbol_data['symbol'] for symbol_data in response.json()]
except requests.exceptions.RequestException as e:
    print(f"Error fetching supported symbols from CoinGecko: {e}")
    supported_symbols = []

# Function to fetch cryptocurrency price from CoinMarketCap API
def get_coinmarketcap_price(api_url, api_key, symbol):
    params = {
        'start': '1',
        'limit': '10',  # Limit the response to 10 items for simplicity
        'convert': 'USD',
    }
    headers = {
        'Accepts': 'application/json',
        'X-CMC_PRO_API_KEY': api_key,
    }
    try:
        response = requests.get(api_url, params=params, headers=headers)
        response.raise_for_status()
        data = response.json()

        if 'data' in data:
            relevant_data = [item for item in data['data'] if item['symbol'] == symbol]
            if relevant_data:
                price = relevant_data[0]['quote']['USD']['price']
                return price
            else:
                logging.warning(f"Symbol {symbol} not found in CoinMarketCap response")
                return None
        else:
            logging.error(f"Error in CoinMarketCap response: {data}")
            return None
    except requests.exceptions.RequestException as e:
        logging.error(f"Error making CoinMarketCap request: {e}")
        return None

# Function to fetch cryptocurrency price from Binance API
def get_binance_price(api_url, api_key, symbol):
    params = {
        'symbol': symbol,
    }
    try:
        response = requests.get(api_url, params=params)
        response.raise_for_status()
        data = response.json()

        if 'price' in data:
            return float(data['price'])
        else:
            logging.warning(f"Symbol {symbol} not found in Binance response")
            return None
    except requests.exceptions.RequestException as e:
        logging.error(f"Error making Binance request: {e}")
        return None

# Function to fetch cryptocurrency price from CoinGecko API
def get_coingecko_price(api_url, symbol):
    params = {
        'ids': symbol.lower(),  # CoinGecko expects lowercase symbol
        'vs_currencies': 'usd',
    }
    try:
        response = requests.get(api_url, params=params)
        response.raise_for_status()
        data = response.json()

        if symbol.lower() in data and 'usd' in data[symbol.lower()]:
            return data[symbol.lower()]['usd']
        else:
            logging.warning(f"Symbol {symbol} not found or no USD price in CoinGecko response")
            return None
    except requests.exceptions.RequestException as e:
        logging.error(f"Error making CoinGecko request: {e}")
        return None

# Function to fetch cryptocurrency prices from all APIs concurrently
def fetch_prices(symbol_coinmarketcap, symbol_binance, symbol_coingecko):
    with ThreadPoolExecutor(max_workers=3) as executor:
        coinmarketcap_future = executor.submit(get_coinmarketcap_price, coinmarketcap_api_url, coinmarketcap_api_key, symbol_coinmarketcap)
        binance_future = executor.submit(get_binance_price, binance_api_url, binance_api_key, symbol_binance)
        coingecko_future = executor.submit(get_coingecko_price, coingecko_api_url, symbol_coingecko)

        price_coinmarketcap = coinmarketcap_future.result()
        price_binance = binance_future.result()
        price_coingecko = coingecko_future.result()

    return price_coinmarketcap, price_binance, price_coingecko

# Prompt the user to choose a cryptocurrency symbol with basic validation
while True:
    symbol_choice = input("Enter the symbol of the cryptocurrency you want to analyze (e.g., BTC, ETH): ").upper()
    if all(symbol.isalpha() and symbol.isupper() for symbol in symbol_choice):
        break
    else:
        print("Invalid symbol. Please enter a valid uppercase alphabetical symbol.")

# Ensure that the chosen symbol is supported by CoinGecko
if symbol_choice not in supported_symbols:
    print(f"Error: Symbol {symbol_choice} is not supported by CoinGecko.")
else:
    # Fetch cryptocurrency prices from all APIs concurrently
    price_coinmarketcap, price_binance, price_coingecko = fetch_prices(symbol_choice, f"{symbol_choice}USDT", symbol_choice)

    # Compare prices
    print(f"Price from CoinMarketCap for {symbol_choice}: {price_coinmarketcap}")
    print(f"Price from Binance for {symbol_choice}: {price_binance}")
    print(f"Price from CoinGecko for {symbol_choice}: {price_coingecko}")

    # Add your comparison logic here
    # For example, find the maximum price among the three
    max_price = max(filter(None, [price_coinmarketcap, price_binance, price_coingecko]))
    print(f"Maximum price among the three: {max_price}")

```
