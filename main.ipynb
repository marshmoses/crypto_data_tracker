```python
import os
import requests
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Define the function to fetch prices
def fetch_prices(symbol, api_key):
    coinmarketcap_api_url = "https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest"
    headers = {
        'X-CMC_PRO_API_KEY': api_key,
        'Accept': 'application/json',
    }
    params = {
        'id': symbol.upper(),  # Ensure the symbol is in uppercase
    }
    response = requests.get(coinmarketcap_api_url, headers=headers, params=params)
    data = response.json()

    # Check if 'data' key is present
    if 'data' in data:
        # Extract the price from the updated structure
        price = data['data'][0]['quote']['USD']['price']
        return price
    else:
        print(f"Unable to retrieve data from CoinMarketCap API. Response: {data}")
        return None

# Retrieve the CoinMarketCap API key from the environment variable
coinmarketcap_api_key = os.environ.get('COINMARKETCAP_API_KEY')

# Check if the API key is available
if coinmarketcap_api_key is None:
    print("CoinMarketCap API key not found. Please set your API key in the environment variable.")
else:
    # Fetch cryptocurrency prices from CoinMarketCap
    symbol_choice = "bitcoin"  # Replace with the desired cryptocurrency symbol (e.g., "bitcoin")
    price_coinmarketcap = fetch_prices(symbol_choice, coinmarketcap_api_key)

    # Print the price
    print(f"Price from CoinMarketCap for {symbol_choice.upper()}: {price_coinmarketcap}")

```
